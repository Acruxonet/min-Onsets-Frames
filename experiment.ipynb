```python
import matplotlib.pyplot as plt
import torch

def plot_mel(mel_db):
    # 注意：你的 mel_db 在 toMel 中做了 transpose(0, 1)
    # 现在的形状应该是 (time, n_mels)
    # 绘图时通常需要转回 (n_mels, time)
    mel_to_show = mel_db.transpose(0, 1).detach().cpu().numpy()

    plt.figure(figsize=(12, 6))
    plt.imshow(mel_to_show, aspect='auto', origin='lower', cmap='magma')
    plt.colorbar(label='Log Power')
    plt.title('Mel Spectrogram Check')
    plt.xlabel('Time Frames')
    plt.ylabel('Mel Bin Index')
    plt.show()
```

```python
import torchcodec
from src.dataset import *

gg = ToMel("data/a1.wav")

print(gg.shape)

plot_mel(gg)
```
